# Utilizando o metodo de monte carlo para calcular o valor aproximado de PI.


**O algoritimo de monte carlo consiste em calcular uma determinada área dificil de ser calculada utilizando numeros gerados aleatoriamente. Nesse caso podemos ver os numeros que forem sorteados e a soma de seus quadrados forem menores ou iguais ao raio ao quadrado da circuferencia que nesse caso é 1. Pode ser considerado que o numero foi sorteado dentro da circunferencia. **

image.png

```python
nin = 0 #numero_dentro circuferencia

for i in range(100000):
    y = np.random.rand()
    x = np.random.rand()
    if x**2 + y**2 <= 1:
        nin += 1

pi = 4 * nin / 1000 #1000 é o numero total de amostras,
```
